```python
import pandas as pd
from metloom.pointdata import CSASMet
from metloom.variables import CSASVariables

station = 'SBSP'  # options are 'SASP', 'SBSP'
#SBSP Senator beck, SASP Swamp angel
site = CSASMet(station)
start_date = pd.to_datetime("2021-03-19")
end_date = pd.to_datetime("2021-03-20")

download_vars = [
    CSASVariables.UP_BROADBAND,
    CSASVariables.UP_NIR_SWIR,
    CSASVariables.DOWN_BROADBAND,
    CSASVariables.DOWN_NIR_SWIR,
    CSASVariables.SNOWDEPTH,
]

df_rad = site.get_hourly_data(start_date, end_date, download_vars)
print(df_rad)
```

```python
from metloom.pointdata import CSASMet, SnotelPointData
from metloom.variables import CSASVariables, SnotelVariables
# Get density from red mountain pass which is near by
red_mountain_site = SnotelPointData(
    "713:CO:SNTL", "Red Mountain Pass"
)
# Download depth and SWE at red mountain
df_swe = red_mountain_site.get_daily_data(
    start_date, end_date, variables=[
        SnotelVariables.SWE, SnotelVariables.SNOWDEPTH
    ]
)
# calculate density (kg/m^3) at the site
df_swe["DENSITY"] = 1000.0 * df_swe["SWE"]/df_swe["SNOWDEPTH"]
```
